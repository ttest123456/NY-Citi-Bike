# Questions ans Answers
1. Q: What is the unique number of station in the dataset?
   A: 844

```python
%%bigquery
SELECT
  COUNT(DISTINCT name) AS unique_name
FROM
  `bigquery-public-data.new_york_citibike.citibike_stations`
```

2. Q: How many Citibikes is in the dataset?
   A: 58937715

```python
%%bigquery
SELECT
  COUNT(*) AS num_citibikes
FROM
  `bigquery-public-data.new_york_citibike.citibike_trips`
```

3. Q: What is the most popular start station and end station?
   A: The most popular start station is Pershing Square North while the most popular end station is E 17 St & Broadway.

```python
%%bigquery
SELECT
  start_station_name AS name,
  start_station_latitude AS lan,
  start_station_longitude AS lon,
  COUNT(*) AS num
FROM
  `bigquery-public-data.new_york_citibike.citibike_trips`
WHERE
  start_station_latitude IS NOT NULL
  AND start_station_latitude!=0
  AND start_station_longitude IS NOT NULL
  AND start_station_latitude!=0
GROUP BY
  name,
  lan,
  lon
ORDER BY
  num DESC
LIMIT 10
```

```python
%%bigquery
SELECT
  end_station_name AS name,
  end_station_latitude AS lan,
  end_station_longitude AS lon,
  COUNT(*) AS num
FROM
  `bigquery-public-data.new_york_citibike.citibike_trips`
WHERE
  end_station_latitude IS NOT NULL
  AND end_station_latitude!=0
  AND end_station_longitude IS NOT NULL
  AND end_station_latitude!=0
GROUP BY
  name,
  lan,
  lon
ORDER BY
  num DESC
LIMIT 10
```

4. Q: What is the gender distribution of the Citibike users?
   A: Male users are more then female users. Male users were 75.8% while female users were 24.2%.

```python
%%bigquery
SELECT
  COUNT(*) AS num,
  gender
FROM
  `bigquery-public-data.new_york_citibike.citibike_trips`
WHERE
  gender!=''
  AND gender!='unknown'
GROUP BY
  gender
ORDER BY
  num
```

5. Q: What is the usertype distribution of the Citibike users?
   A: There are a lot more Subscribers then Customers. Subscriber users were 88.3% while customer users were 11.7%.

```python
%%bigquery
SELECT
  COUNT(*) AS num,
  usertype
FROM
  `bigquery-public-data.new_york_citibike.citibike_trips`
WHERE
  usertype!=''
GROUP BY
  usertype
ORDER BY
  num
```

6. Q: What is the trip duration distribution of the Citibike trips?

```python
%%bigquery
SELECT
  COUNT(*) AS num,
  ROUND(SUM(CAST(tripduration AS INT64))/60,2) AS duration,
  CONCAT(start_station_name, 'to', end_station_name) AS trip
FROM
  `bigquery-public-data.new_york_citibike.citibike_trips`
WHERE
  tripduration IS NOT NULL
GROUP BY
  start_station_name,
  end_station_name
ORDER BY
  duration 
LIMIT 10 
```

7. Q: What is the most popular Citibike trip?
   A: The most popular Citibike trip is Central Park S & 6 Ave to Central Park S & 6 Ave.

```python
%%bigquery
SELECT
  COUNT(*) AS num,
  CONCAT(start_station_name, 'to', end_station_name) AS trip
FROM
  `bigquery-public-data.new_york_citibike.citibike_trips`
WHERE
  start_station_name!=''
  AND end_station_name!=''
GROUP BY
  start_station_name,
  end_station_name
ORDER BY
  num DESC
LIMIT 10
```

8. Q: Was there new bike introduced or removed at any point of time? What about station?

```python
%%bigquery 
SELECT
  DISTINCT name AS unique_name,
  CASE
    WHEN SUM(capacity) - (SUM(num_docks_available) + SUM( num_docks_disabled)+SUM(num_bikes_available) + SUM( num_bikes_disabled))>0 THEN 'possible_introduced'
    WHEN SUM(capacity) - (SUM(num_docks_available) + SUM( num_docks_disabled)+SUM(num_bikes_available) + SUM( num_bikes_disabled))<0 THEN 'possible_removed'
  ELSE
  'nothing'
END
  AS introduced_removed
FROM
  `bigquery-public-data.new_york_citibike.citibike_stations`
GROUP BY
  unique_name
ORDER BY
  unique_name
LIMIT 10000
```
